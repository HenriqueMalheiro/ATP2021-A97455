### Resumo: 
    
    Inicialmente, o programa imprime as regras do jogo, explicando também os comandos que devem ser usados durante o mesmo. 
    De seguida, cria-se três variáveis, a inicio de valor 0 que determina o menor valor do intervalo, a variável fim de valor 100 que determina o maior valor do intervalo e a variável contagem de valor 0 que vai ser utilizada na contagem das tentativas do computador para encontrar o numero, que neste caso não poderá ser maior que 7 pois 2^7=128. Estas três variáveis serão usadas na função pesquisabinaria(inicio,fim,contagem).
    Dentro da função, criamos mais uma variável que através da divisão inteira determinará o meio do intervalo, no caso da soma dos extremos ser par, ou o meio de menor valor, no caso da soma dos extremos ser ímpar. Depois, chama-mos "a" uma variável que tomará o valor do input que pergunta se o numero que o utilizador pensou foi o numero a meio do intervalo.
    Assim, vamos ter 3 hipóteses:
    - O valor pensado ser maior do que o valor médio do intervalo;
    - O valor pensado ser menor do que o valor médio do intervalo;
    - O valor pensado ser o valor médio do intervalo.
    Se o valor pensado for maior que o valor médio, a variável inicio tomará o valor do valor médio, tornando o intervalo na parte superior do intervalo anterior. Adiciona-se um à contagem de tentativas, chamando-se de seguida a função de forma recursiva, repetindo o processo.
    Se o valor pensado for menor que o valor médio, a variável fim tomará o valor do valor médio, tornando o intervalo na parte inferior do intervalo anterior. Adiciona-se um à contagem de tentativas, chamando-se de seguida a função de forma recursiva, repetindo o processo.
    Quando o valor médio do intervalo for igual ao valor pensado, adiciona-se um à contagem, imprimindo-se o número de tentativas da máquina para adivinhar, tendo a particularidade de se a contagem for 1 imprime-se no singular, caso contrário imprime-se no plural.
