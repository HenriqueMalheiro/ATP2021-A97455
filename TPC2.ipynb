```python
print("Pense num numero de 0 a 100")
print("Digite + se for maior que o numero apresentado")
print("Digite - se for menor que o numero apresentado")
print("Digite sim se for o numero em que pensou")

inicio=0
fim=100
contagem=0

def pesquisabinaria(inicio,fim,contagem):
    meio=(inicio+fim)//2
    a=input("E o "+str(meio)+"? ")
    if a=="sim":
        contagem=contagem+1
        if contagem==1:
            print("Numero encontrado em",str(contagem),"tentativa")
        else:
            print("Numero encontrado em",str(contagem),"tentativas")
    elif a=="+":
        inicio=meio
        contagem=contagem+1
        pesquisabinaria(inicio,fim,contagem)
    elif a=="-":
        contagem=contagem+1
        fim=meio
        pesquisabinaria(inicio,fim,contagem)
    else:
        print("Digite um dos caracteres permitidos!")
        
    return "Obrigado por ter jogado!"

print(pesquisabinaria(inicio,fim,contagem))
```
