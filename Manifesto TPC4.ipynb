### Resumo:

Inicialmente, mostra-se o menu que vai ter 10 opções diferentes para o utilizador escolher, sendo o mesmo corrido até a opção 0 ser escolhida(opção sair).
As funções subtrair e multiplicar fazem o mesmo processo que a função soma, com a diferença do cálculo da variável a, que se adequa à operação matemática em questão. Já a função dividir tem uma pequena particularidade, pois simplesmente faz o inverso da fração do divisor, chamando de seguida a função multiplicar, pois dividir é a mesma coisa que multiplicar pelo inverso.
A função 7, pergunta inicialmente o numero de frações na lista criada aleatoriamente. A funçao vai gerar dois valores aleatorios, criando um tuplo que de seguida é adicionado a uma lista inicialmente vazia.
As funções 8 e 9 dependem da 7 na medida em que vão utilizar a lista de funções criada para a soma ou para verificar o maior.
Na soma, corre-se todas as frações da lista e adiciona-se a uma variavel res, sendo esse res final a soma de todas as frações da lista.
Na maior fração, toma-se como a maior a primeira fração da lista. A variavel maior2 tomara o valor dessa fracao e vai compara-lo com o valor das fracoes seguintes(div). Se o div for maior que o maior2, atualiza-se os valores de maior2 e maior, imprimindo-se o valor que ficar em maior(maior fraçao da lista).
