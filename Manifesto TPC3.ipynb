### Resumo:

Se o utilizador escolher a opção 4, irá ser chamada a função bubblesort(l) que utiliza a lista criada anteriormente, de maneira manual ou aleatoria, e a vai ordenar do menor para o maior número.
Na função, começamos por definir a variável troca de valor 1. 
Depois, impomos um ciclo while que vai correr até a variável troca não ser 1. Dentro do while, temos um ciclo for que vai dar a i todos os valores de 0 até ao (tamanho da lista-1), ou seja, todos os índices da lista com a excessão do último elemento.
De seguida, se um dos valores for maior do que o valor seguinte, cria-se duas variáveis "j" e "k" que tomam os seus valores e os trocam de posição(entende-se assim o porquê de se desprezar o ultimo valor no ciclo for, pois o mesmo vai ser comparado com o penultimo elemento). Chama-se depois a função de maneira recursiva, repetindo o processo.
Quando a lista estiver ordenada, o ciclo if não irá ser acionado, entrando no else que toma o valor da variável troca como 0, o que permite encerrar o ciclo while.
Por fim, retorna-se a lista ordenada, imprimindo-se de seguida para ser vista pelo utilizador.
