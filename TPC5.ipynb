```python
# Construtor

def criarTermo(coef,exp):
    return (coef, exp)


def criarPolinomio():
    return []


def inserirTermo(p,t):
    return p.append(t)

# Ver Polinomio

def verTermo(t):
    c, e = t
    if e == 0:
        return str(c)
    elif e == 1:
        return str(c)+"x"
    return str(c)+"x^"+str(e)


def verPolinomio(p):
    res = ""
    for t in p:
        if res == "":
            res = verTermo(t)
        else:
            res = res + " " + "+" + " " + verTermo(t)
    print(res)

# Calcular o polinomio

def calcPol(p,x):
    res = 0
    for t in p:
        c, e = t
        valor = (x**e)*c
        res = res+valor
    return res

# Tabela de valores

def tabela(p,nlinhas):
    for i in range(nlinhas):
        print(str(i) + "::" + str(calcPol(p, i)))

# Ordenar o polinomio

def chaveOrd(t):
    c,e=t
    return e


def ordenarPol(p):
    p.sort(reverse=True,key=chaveOrd)
    return p

# Simplificar o polinomio

def simplificarPol(p):
    p = ordenarPol(p)
    p2 = criarPolinomio()
    for t in p:
        t2=[t[0],t[1]]
        inserirTermo(p2,t2)
    i=0
    while i<len(p2)-1:
        if p2[i][1]==p2[i+1][1]:
            c=p2[i][0]+p2[i+1][0]
            t=criarTermo(c,p2[i][1])
            inserirTermo(p2,t)
            p2.pop(i)
            p2.pop(i)
            p2=ordenarPol(p2)
            i=0
        else:
            i=i+1
    return p2

# Derivar o polinomio

def derivarPol(p):
    res = []
    for t in p:
        c, e = t
        if e > 0:
            res.append((c*e, e-1))
    return res

# Menu

def menu():
    print(""" 
1- criar polinomio
2- calcular o polinomio
3- fazer uma tabela
4- simplificar o polinomio
5- derivar o polinomio
0- sair""")

# Executar o menu

opcao="1"
while opcao!="0":
    menu()
    opcao=input("Qual a sua opção?")
    if opcao=="1":
        tamanho=int(input("Quantos termos tem o polinomio?"))
        p=criarPolinomio()
        for i in range(tamanho):
            c=int(input("Diga o coeficiente:"))
            e=int(input("Diga o expoente:"))
            t=criarTermo(c,e)
            inserirTermo(p,t)
        verPolinomio(p)
    elif opcao=="2":
        x=int(input("Diga o valor de x:"))
        print(calcPol(p,x))
    elif opcao=="3":
        nlinhas=int(input("Diga o nº de linhas da tabela:"))
        tabela(p,nlinhas)
    elif opcao=="4":
        verPolinomio(simplificarPol(p))
    elif opcao=="5":
        p=simplificarPol(p)
        verPolinomio(derivarPol(p))
    elif opcao=="0":
        print("Obrigado por usar o programa")
    else:
        print("Opcao invalida")
```
