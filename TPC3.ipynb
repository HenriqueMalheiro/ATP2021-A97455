```python
# Criar uma lista à mão
def listam():
    l=[]
    a=int(input("Nº de elementos da lista:"))
    for i in range(a):
        b=input("Diga os elementos:")
        l.append(b)
    return l

# Cria uma lista aleatoria
def listaa():
    import random
    l=[]
    a=int(input("Nº de elementos da lista:"))
    for i in range(a):
        b=random.randint(0,100)
        l.append(b)
    return l

# Menu

def menu():
    print("""Gestão de listas de inteiros
1) Criar lista à mão
2) Criar lista aleatoriamente
3) Maior elemento da lista
4) Ordenar a lista
0) Sair""")
    
    
# Maior elemento 

def maiorelemento(l):
    maior=l[0]
    for i in range(len(l)):
        if l[i]>maior:
            maior=l[i]
    return maior

# Ordenar a lista

def bubblesort(l):
    troca=1
    while troca==1:
        for i in range(len(l)-1):
            if l[i]>l[i+1]:
                j=l[i+1]
                k=l[i]
                l[i]=j
                l[i+1]=k
                bubblesort(l)
            else:
                troca=0
    return l
    
# Executar o menu

opcao="1"
while opcao!="0":
    menu()
    opcao=input("Qual a sua opção?")
    if opcao == "1":
        l=listam()
        print(l)
    elif opcao== "2":
        l=listaa()
        print(l)
    elif opcao=="0":
        print("Ate à proxima")
    elif opcao=="3":
        print(maiorelemento(l))
    elif opcao=="4":
        print(bubblesort(l))
    else:
        print("Opção invàlida")
```
