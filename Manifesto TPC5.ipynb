### Resumo:

   Para simplificar o polinomio, precisamos de ter a função ordenada pelos expoentes dos termos e transformar os termos do polinomio em listas ao inves dos tuplos, de maneira a podermos trabalhar com os valores dos seus coeficientes.
   Assim, criamos um segundo polinomio(p2) que vai adicionar todos os termos do polinomio p atraves do ciclo "for" estabelecido.
   Comecamos um ciclo "while" em que vamos comparar o expoente do primeiro termo com o seguinte(len(p)-1, pois o ultimo indice nao precisa ser comparado visto que ja tera sido comparado com o penultimo). 
Se forem iguais, soma os coeficientes e adiciona este termo soma ao polinomio, retirando os dois termos utilizados na soma, voltando ao indice 0 voltando a percorrer a partir do primeiro termo.
Se forem diferentes, soma um ao indice e ira comparar o segundo com o terceiro termo, e dai sucessivamente em toda a extensao do polinomio.
   No fim retorna-se o polinomio p2, sendo impresso atraves da funcao verPolinomio() no menu.
